# Relatório Operacional de Pré-Decolagem — Aurora Seeger


## 1. Dados de Telemetria

```python
telemetria = {
    "temperatura_interna": 22.5,   # °C
    "temperatura_externa": -60.0,  # °C
    "integridade_estrutural": 1,   # 1 = OK, 0 = falha
    "nivel_energia": 87.3,         # %
    "pressao_tanque": 350.0,        # kPa
    "modulo_propulsao": 1,
    "modulo_comunicacao": 1,
    "modulo_navegacao": 1,
}
```

## 2. Algoritmo de Verificação
[fluxograma em texto/pseudocódigo + código Python]

```python
def verificar_decolagem(t):
    falhas = []

    if not (18 <= t["temperatura_interna"] <= 30):
        falhas.append("Temperatura interna fora do range")
    if t["integridade_estrutural"] != 1:
        falhas.append("Falha na integridade estrutural")
    if t["nivel_energia"] < 80:
        falhas.append("Energia abaixo do mínimo")
    if not (300 <= t["pressao_tanque"] <= 400):
        falhas.append("Pressão do tanque fora do range")
    for modulo in ["modulo_propulsao", "modulo_comunicacao", "modulo_navegacao"]:
        if t[modulo] != 1:
            falhas.append(f"Módulo com falha: {modulo}")

    if not falhas:
        print("✅ STATUS: PRONTO PARA DECOLAR")
    else:
        print("❌ STATUS: DECOLAGEM ABORTADA")
        for f in falhas:
            print(f"  → {f}")

verificar_decolagem(telemetria)
```

## 3. Análise Energética

```python
capacidade_total_kwh = 500
carga_atual_pct = telemetria["nivel_energia"]
consumo_decolagem_kwh = 120
perda_estimada_pct = 5  # perda em %

energia_disponivel = capacidade_total_kwh * (carga_atual_pct / 100)
energia_apos_perda = energia_disponivel * (1 - perda_estimada_pct / 100)
autonomia_restante = energia_apos_perda - consumo_decolagem_kwh

print(f"Energia disponível: {energia_disponivel:.1f} kWh")
print(f"Após perdas: {energia_apos_perda:.1f} kWh")
print(f"Após decolagem (reserva): {autonomia_restante:.1f} kWh")
```
